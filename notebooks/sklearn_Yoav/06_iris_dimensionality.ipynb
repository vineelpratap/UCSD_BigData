## The Math of PCA ###

We start with PCA in 2D, which is convinient because we can plot everything. Performing PCA in 2D corrsponds to identifying the rotation 
and scaling of the data so that the coordinates become uncorrelated with variance of 1 each.

The matrix that describes a 2D rotation around the origin with angle $\theta$ is

<img src="files/Rotation_matrix.jpg"/>

```python
from random import uniform
from numpy.linalg import norm
n=1000
i=0
shift=np.matrix([[3],[5]])
theta=0.5
rotation=np.matrix([[cos(theta),-sin(theta)],[sin(theta),cos(theta)]])
scale=np.matrix([1,0.1]).transpose()
print 'rotation matrix=\n',mat
print 'scaling factors=',scale.transpose()

X=np.zeros([2,n])
while i<n:
    p=np.matrix([uniform(-1,1),uniform(-1,1)]).transpose()
    if norm(p)<1:
        p=rotation*np.multiply(p,scale)
        X[:,i]=np.squeeze(np.asarray(p+shift))
        #list(p*mat)
        i+=1
scatter(X[0,:],X[1,:])
#xlim([0,10])
#ylim([0,10])
grid()
```

## Dimensionality Reduction: PCA

Dimensionality reduction is the task of deriving a set of new
artificial features that is smaller than the original feature
set while retaining most of the variance of the original data.
Here we'll use a common but powerful dimensionality reduction
technique called Principal Component Analysis (PCA).
We'll perform PCA on the iris dataset that we saw before:

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
print shape(X),shape(y)
```

PCA is performed using linear combinations of the original features
using a truncated Singular Value Decomposition of the matrix X so
as to project the data onto a base of the top singular vectors.
If the number of retained components is 2 or 3, PCA can be used
to visualize the dataset.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, whiten=True)
pca.fit(X)
```

Once fitted, the pca model exposes the singular vectors in the components_ attribute:

```python
pca.components_
```

Other attributes are available as well:

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

Let us project the iris dataset along those first two dimensions:

```python
X_pca = pca.transform(X)
print shape(X), shape(X_pca)
```

PCA `normalizes` and `whitens` the data, which means that the data
is now centered on both components with unit variance:

```python
import numpy as np
np.round(X_pca.mean(axis=0), decimals=5)
```

```python
np.round(X_pca.std(axis=0), decimals=5)
```

Furthermore, the samples components do no longer carry any linear correlation:

```python
np.corrcoef(X_pca.T)
```

We can visualize the projection using pylab, but first
let's make sure our ipython notebook is in pylab inline mode

```python
%pylab inline
```

Now we can visualize the results using the following utility function:

```python
import pylab as pl
from itertools import cycle

def plot_PCA_2D(data, target, target_names, title = ""):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
    pl.legend()
    pl.title(title)
```

Now calling this function for our data, we see the plot:

```python
plot_PCA_2D(X_pca, iris.target, iris.target_names, "PCA of IRIS dataset")
```

Note that this projection was determined *without* any information about the
labels (represented by the colors): this is the sense in which the learning
is **unsupervised**.  Nevertheless, we see that the projection gives us insight
into the distribution of the different flowers in parameter space: notably,
*iris setosa* is much more distinct than the other two species.

Note also that the default implementation of PCA computes the
singular value decomposition (SVD) of the full
data matrix, which is not scalable when both ``n_samples`` and
``n_features`` are big (more that a few thousands).
If you are interested in a number of components that is much
smaller than both ``n_samples`` and ``n_features``, consider using
`sklearn.decomposition.RandomizedPCA` instead.

Other dimensionality reduction techniques which are useful to know about:

- `sklearn.decomposition.PCA`: <br/>
   Principal Component Analysis
- `sklearn.decomposition.RandomizedPCA`:<br/>
   fast non-exact PCA implementation based on a randomized algorithm
- `sklearn.decomposition.SparsePCA`:<br/>
   PCA variant including L1 penalty for sparsity
- `sklearn.decomposition.FastICA`:<br/>
   Independent Component Analysis
- `sklearn.decomposition.NMF`:<br/>
   non-negative matrix factorization
- `sklearn.manifold.LocallyLinearEmbedding`: <br/>
   nonlinear manifold learning technique based on local neighborhood geometry
- `sklearn.manifold.IsoMap`: <br/>
   nonlinear manifold learning technique based on a sparse graph algorithm

##Dimensionality reduction using Randomized PCA##
Now we'll use Randomized PCA to perform dimensionality reduction on Iris dataset. Randomized PCA  uses approximated Singular Value Decomposition of the data that preserves most of the variance, by dropping the singular vector of components associated with lower singular values. (http://scikit-learn.org/stable/modules/decomposition.html)

Loading modules and fitting the model with Iris data set. <br/>
No. of components chosen for the Randomized PCA model = 2

```python
from sklearn.decomposition import RandomizedPCA
rand_pca = RandomizedPCA(n_components=2, whiten=True)
rand_pca.fit(X)
```

Comparing first two components of maximum variance using PCA and Randomized PCA

```python
print "Randomized PCA components "
print rand_pca.components_ 
print 
print "PCA components "
print pca.components_
```

We can observe that the first principal component of randomized PCA is the same as PCA while the second component of randomized PCA is the negative of PCA's second component. 

Comparing the % of variance explained by each of the two selected components. 

```python
print "Randomized PCA explained variance ratio"
print rand_pca.explained_variance_ratio_
print 
print "PCA explained variance ratio"
print pca.explained_variance_ratio_
```

Calculating the sum of the explained variance ratio. 

```python
print "Sum of explained variance ratio by first 2 components"
print "Randomized PCA : " , rand_pca.explained_variance_ratio_.sum()
print "PCA : ", pca.explained_variance_ratio_.sum()
```

Though the first two component is the same, percentage of variance explained by it is different for the two methods. This is because Randomized PCA calculates only the required number of singular vectors and divides the variance among them. Thus Randomized PCA can be performed faster with less memory.  <br />

### Deducing true explained variance ###

From the above discussion, it can be seen that the sum of explained variance ratio given by Randomized PCA will be always 1. As this is clearly not the case, we will find the true explained variance w.r.t the original data here. <br />

We first calculate the total variance present in the given data

```python
print "Variance along each component for the original data :", np.var(X, axis = 0) 
print 
tot_var = sum(np.var(X, axis = 0))
print "Total variance in data", np.var(X, axis = 0).sum()
```

```python
print "Variance along top k (= 2) components for the tranformed data :", rand_pca.explained_variance_
print 
rand_tot_var = rand_pca.explained_variance_.sum()
print "Total variance in the tranformed data", rand_tot_var.sum()
```

Now, we take the ratio w.r.t to the original data to find the percentage of total varaince explained

```python
explained_variance_ratio = rand_pca.explained_variance_/tot_var
print "explained variance ratio along each component", explained_variance_ratio
print
print "Total explained variance :", explained_variance_ratio.sum()
print 
print "[Sanity Check] Total explained variance given by PCA:", pca.explained_variance_ratio_.sum()
```

Applying dimensionality reduction on the dataset

```python
X_rand_pca = rand_pca.transform(X)
print shape(X), shape(X_rand_pca)
```

Similar to PCA, Randomized PCA normalizes and whitens the data.

```python
print "mean : ", np.round(X_rand_pca.mean(axis=0), decimals=5)
print "standard deviation : ", np.round(X_rand_pca.std(axis=0), decimals=5)
```

Also, no correlation exists between the components after projecting them to lower dimensional space. 

```python
np.corrcoef(X_rand_pca.T)
```

Comparing projections of PCA and Randomized PCA

```python
plot_PCA_2D(X_rand_pca, iris.target, iris.target_names, "Randomized PCA of IRIS dataset")
```

```python
plot_PCA_2D(X_pca, iris.target, iris.target_names,"PCA of IRIS dataset" )
```

Due to the second principal component of Randomized PCA being negative of PCA, its projection is a flipped version of projection of PCA across the horizontal axis.
